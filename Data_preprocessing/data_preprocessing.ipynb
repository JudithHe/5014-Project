```python
#split the data into 4 parts
nation_split = np.array_split(nation, 4)
nation25, nation50, nation75, nation100=0,0,0,0
df = [nation25, nation50, nation75, nation100]
for i in range(4):
    df[i] = pd.DataFrame(nation_split[i]) 
    tag = [str(25*(i+1))]*len(df[i])
    table = ['nation']*len(df[i])
    df[i]['tag'] = tag
    df[i]['table'] = table
#combine them together
#nation_new = pd.concat(df)
#display(nation_new)
```
